```python
data = df.to_dict(orient='records')
data
```

```python
DB_NAME = "US_VISA_DB"
COLLECTION_NAME = "VISA_DATA_COLLECTION"
CONNECTION_URL = ""
```

```python
from pymongo import MongoClient
```

```python
client = MongoClient(CONNECTION_URL)
db = client[DB_NAME]
collection = db[COLLECTION_NAME]
```

```python
records = collection.insert_many(data)
records
```

```python
recs = collection.find()
type(recs)
```
